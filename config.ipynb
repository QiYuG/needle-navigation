# 文件&函数作用
## utility.py：用于实现一些函数
### calculate_loss
- 损失计算函数
- 输入：
  - heart_vertices: 心脏表面顶点坐标数组 (N,3)
  - catheter_line: 导管线坐标数组 (M,3)
  - threshold: 距离阈值，超过该值的点不计入计算
- 输出：
  - total_distance: 惩罚后的最大距离值
### calculate_min_dis
- 最小距离计算函数
- 输入：
    - heart_vertices: 心脏表面顶点坐标数组 (N,3)
    - catheter_line: 导管线坐标数组 (M,3)
- 输出：
    - min_distance: 最小欧氏距离
    - id: 最小距离对应的顶点和导管点索引元组 (vertex_idx,   catheter_point_idx)
### fit_plane_and_find_normal_line
- 平面拟合与法线计算函数
- 输入：
    - points: 输入点云坐标数组 (N,3)
    - threshold: 距离阈值，用于筛选有效点
    - modify: 质心位置调整偏移量
- 输出：
    - centroid: 调整后的质心坐标
    - normal_vector: 平面法线向量
### angle_between_vectors
- 向量夹角计算函数
### solve_parameters
- 几何参数求解函数
- 输入：
    - centroid: 平面质心坐标
    - normal_vector: 平面法线向量
    - theta_x: 输入角度（度）
    - r_x: 输入半径参数
- 输出：
    - theta_z: 求解的z轴旋转角度（弧度）
    - d_z: z轴方向偏移量
    - t: 直线参数方程中的比例系数
## Needle.py：用于定义机器人对象及其运动学
### forward_kinematics
正运动学计算函数，得到机器人某电脑的位姿
### calculate_shape
计算导管的三维形状，以点云形式储存在calculate_shape属性（全部点云）、catheter_bending_section属性（弯曲段）与catheter_rigid_section属性（刚性段）中
### transformation_matrix_z(theta, d)
Z轴旋转平移矩阵， 对应机器人绕轴自转与沿轴伸长
### constance_curve_matrix_x(theta, r)
恒定曲率弯曲矩阵（绕X轴），对应机器人整体相对底座的弯曲与伸长
### transl(x,y,z)
纯平移矩阵，对应机器人底座的平移
### transformation_matrix_y(theta, d)
Y轴旋转平移矩阵，对应机器人弯曲，移动没理解
